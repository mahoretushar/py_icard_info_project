# Information Extractor from Hall Ticket

This code extracts information from hall ticket's of students generated by the SGBAU. Hall ticket's must be provided in PDF format, the standard format provided by the University. Please take reference of the 3rd_ht.pdf present in this directory. If the PDF is not of the same format as that of the provided sample then the pattern matching in following code may not work.

### Libraries

Following block of code imports required libraries, make sure that the libraries are installed properly, if not then make use of "pip" and install them first. (example: pip install PyPDF2).

- PyPDF2 is used for the pdf file handeling
- re is used for the pattern matching in the form of regular expression
- csv for converting the the output in .csv file for further use

```python
import PyPDF2
import re
import csv
```

### File Handeling Object and Opening a File

Following block of code creates a file handeling object and opens a file for processing. At last for confirmation printing out the total number of pagse by using ".numPages" method.

```python
pdfFileObject = open('3rd_ht.pdf','rb')
pdfReaderObject = PyPDF2.PdfFileReader(pdfFileObject)
print(pdfReaderObject.numPages)
```

## Geting Name of Student

Following code block shows the logic of how to get the Name of Student from the first page of the PDF. After observation of the provided PDF you can see that every page contains information of single student on each page. So the above count shows total pages that means total students are also 67. 

Now firstly extract the text of first page and observe the output, observing the output will help in understanding how to apply regular expression for pattern matching to get the required output.

```python
firstPageObject = pdfReaderObject.getPage(0)
text = firstPageObject.extractText()
print(text)
```

### Continued...

After observing the above output you can see that the name of the student is at the begining of the text, so to extract it we can simply split the text for the "Name" word and after that if we print index[0] we will get the required output.

```python
text1 = text.split("Name")
text2 = text1[0]
#text2 = text2.split(" ")
print(text2)
```

### Continued...

Further more we can delete the remaning part and get just the name of the student, the following line of codes will show how to do it.

```python
del text1[1:4]
print(text1)
```

## Main Logic

Now we can apply the above logic on the complete PDF file by using for loop, we need to use the ".pages" method for itteration along with the reader object we have created earlier. The rest of the code is selfexplanatory.

```python
name_list = []
for page in pdfReaderObject.pages:
    text = page.extractText()
    text1 = text.split("Name")
    del text1[1:4]
    name_list.append(list(text1))
```

## Logic for Roll No.

To get the roll no. of the student we have to use regular expression for the pattern matching, again observation is needed. First of all we have to identify where the Roll No. is present, then we need to select the pattern, remember the pattern must be unique (because we need single roll no. from single page). After observation "71\d{3}809" is what we can use for identification pattern. Following code shows the single run, after that this logic is implimented on compelete document.

To implement this logic first of all we need to store the text of whole document in a list variable "all_text" and the appply the pattern matching by using ".findall" method.

```python
pattern = r"71\d{3}809"
match = re.search(pattern,text)
print(match)
```

```python
# Extracting text from whole document
all_text = []
for page in pdfReaderObject.pages:
    text = page.extractText()
    text = text.split("\n")
    all_text.append(list(text))
```

```python
# Appling pattern matching on all text, and getting the output
pattern = r"71\d{3}"
match = re.findall(pattern,str(all_text))
print(match)
```

```python
# Converting the list into nested list, its needed

new_list = []

# Creating list of list format
for elem in match:
   temp = elem.split(', ')
   new_list.append((temp))

# Final list
res = []

for elem in new_list:
   temp = []
   for e in elem:
      temp.append(e)
   res.append(temp)

# printing
print("The list of lists:\n",res)
```

## Converting to CSV

Following code block converts the output into the .csv file which can be further used for various purpose

```python
import csv
fields = ["Name","Roll No."]
with open('name','w') as f:
    write = csv.writer(f)
    write.writerows(name_list)    
```
